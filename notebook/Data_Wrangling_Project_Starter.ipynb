#### Dataset 2

Type: *HTML Table scraped from Wikipedia.*

Method: *The dataset was gathered by extracting a table from an HTML page using `BeautifulSoup` and `pandas.read_html`.*

Dataset variables:

*   *`Entity`: Country name.*
*   *`PM2.5 air pollution (micrograms per cubic meter)`: Fine particulate pollution level.*

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

# Step 1: Get the Wikipedia page
url = 'https://en.wikipedia.org/wiki/List_of_countries_by_air_pollution'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
tables = soup.find_all('table', {'class': 'wikitable'})
df_html_pollution = pd.read_html(str(tables[0]))[0]

df_html_pollution = df_html_pollution[['Country', 'Particulate matter (PM2.5)  (in μg/m3) WHO-Guideline=5']]
df_html_pollution.columns = ['Country', 'PM2.5']

df_html_pollution.head()
```

## 3. Clean data
Clean the data to solve the 4 issues corresponding to data quality and tidiness found in the assessing step. **Make sure you include justifications for your cleaning decisions.**

After the cleaning for each issue, please use **either** the visually or programatical method to validate the cleaning was succesful.

At this stage, you are also expected to remove variables that are unnecessary for your analysis and combine your datasets. Depending on your datasets, you may choose to perform variable combination and elimination before or after the cleaning stage. Your dataset must have **at least** 4 variables after combining the data.

```python
happiness_clean = happiness_df.copy()
pollution_clean = df_html_pollution.copy()
```

### **Quality Issue 1: Remove rows with PM2.5 = 0**

```python
pollution_clean = pollution_clean[pollution_clean['PM2.5'] > 0]
```

Justification: *Zero values in the 'Value PM25' column likely represent missing data rather than actual zero pollution. Including them would bias the national average downward and distort the relationship between air quality and happiness. Removing these entries ensures more accurate and meaningful analysis.*

### **Quality Issue 2: Aggregate pollution data by country**

```python
pollution_country_avg = pollution_clean.groupby('Country')['PM2.5'].mean().reset_index()
pollution_country_avg.columns = ['Country', 'PM2.5']

pollution_country_avg.head()
```

Justification: *To perform country-level analysis that aligns with the happiness dataset, I aggregate the data by computing the mean PM2.5 value for each country with grouping pollution data by `Country` and compute the average PM2.5 value to ensure one row per country matching the happiness dataset structure.*

### **Tidiness Issue 1: Observational unit mismatch**

```python
pollution_country_avg = pollution_clean.groupby('Country')['PM2.5'].mean().reset_index()
pollution_country_avg.columns = ['Country', 'PM2.5']

pollution_country_avg.head()
```

Justification: *In the pollution dataset, each row represents pollution data for a specific city, while the happiness dataset is aggregated at the country level. This violates the rule that "each type of observational unit forms a table." To align the data, the pollution values must be aggregated by country before merging.*

### **Tidiness Issue 2: Standardize column names**

```python
happiness_clean.columns = [col.strip().replace(' ', '_') for col in happiness_clean.columns]
```

Justification: *Column names containing spaces can complicate code syntax and cause issues during data manipulation. Replacing spaces with underscores improves readability.*

### **Remove unnecessary variables and combine datasets**

Depending on the datasets, you can also peform the combination before the cleaning steps.

```python
merged_df = pd.merge(happiness_clean, pollution_country_avg, how='inner', left_on='Country_name', right_on='Country')
merged_df.drop(columns=['Country'], inplace=True)

merged_df.head()
```

## 5. Answer the research question

### **5.1:** Define and answer the research question
Going back to the problem statement in step 1, use the cleaned data to answer the question you raised. Produce **at least** two visualizations using the cleaned data and explain how they help you answer the question.

*Research question:* Does a higher level of PM2.5 air pollution correlate with lower happiness scores at the national level?

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.scatterplot(data=merged_df, x='PM2.5', y='Ladder_score')
plt.title('Happiness Score vs. PM2.5 Air Pollution by Country')
plt.xlabel('Average PM2.5 Level')
plt.ylabel('Happiness Score')
plt.grid(True)
plt.show()
```

*Answer to research question:*
The scatter plot suggests a potential negative trend — countries with higher PM2.5 pollution levels often report lower happiness scores.

```python
plt.figure(figsize=(6, 4))
sns.heatmap(merged_df[['Ladder_score', 'PM2.5']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation between Happiness and Pollution')
plt.show()
```

*Answer to research question:* The correlation value confirms a moderate negative correlation between pollution and happiness scores.
